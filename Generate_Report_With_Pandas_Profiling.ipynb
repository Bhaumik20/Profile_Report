```python
import pandas as pd
```

```python
branch_adm = pd.read_csv('csv/d_branch_adm.csv')
branch_adm.set_index('BRANCH_ID',inplace=True)
branch_adm.head()
```

```python
fellow_adm = pd.read_csv('csv/d_fellow_adm.csv')
fellow_adm.set_index('FELLOW_ID',inplace=True)
fellow_adm.head()
```

```python
period_adm = pd.read_csv('csv/d_period_adm.csv')
period_adm.set_index('PERIOD_ID',inplace=True)
period_adm.head()
```

```python
f_adm = pd.read_csv('csv/f_adm.csv')
f_adm.head()
```

```python
f_adm.info()
```

```python
f_adm['BRANCH_STREAM']=f_adm['BRANCH_ID'].apply(lambda x : branch_adm.loc[x]['BRANCH_STREAM'])
f_adm['BRANCH_CATEGORY']=f_adm['BRANCH_ID'].apply(lambda x : branch_adm.loc[x]['BRANCH_CATEGORY'])
```

```python
f_adm['FELLOW_GENDER']=f_adm['FELLOW_ID'].apply(lambda x : fellow_adm.loc[x]['FELLOW_GENDER'])
f_adm['FELLOW_CATEGORY']=f_adm['FELLOW_ID'].apply(lambda x : fellow_adm.loc[x]['FELLOW_CATEGORY'])
```

```python
f_adm.head()
```

```python
f_adm['PERIOD_YEAR']=f_adm['PERIOD_ID'].apply(lambda x : period_adm.loc[x]['PERIOD_YEAR'])
f_adm['PERIOD_TYPE']=f_adm['PERIOD_ID'].apply(lambda x : period_adm.loc[x]['PERIOD_TYPE'])
```

```python
f_adm.head()
```

```python
df = f_adm.drop(['BRANCH_ID','FELLOW_ID','PERIOD_ID'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
from pandas_profiling import ProfileReport
```

```python
prof = ProfileReport(df)
prof.to_file(outputfile='output.html')
```
